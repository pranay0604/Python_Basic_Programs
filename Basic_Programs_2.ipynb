### Level 2: Number-based Looping Logic

```python
# 1. Count the number of digits in a given number.

n = int(input("Enter the number"))
count = 0
temp = n
while temp>0:
    digit = temp%10
    count += 1
    temp //= 10
print(count)
```

```python
# 2. Print the reverse of a given number.

n = int(input("Enter the number"))
temp = n
rev = 0
while temp>0:
    digit = temp%10
    rev = rev * 10 + digit
    temp //= 10

print(rev)    
```

```python
# 3. Check if a number is a palindrome.

n = int(input("Enter the number"))
temp = n
pal = 0

while temp>0:
    digit = temp%10
    pal = pal * 10 + digit
    temp //= 10

if n == pal:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
# 4. Find the sum of digits of a number.

n = int(input("Enter the number"))
summ = 0
temp = n
while temp>0:
    digit = temp%10
    summ += digit
    temp //= 10
print(summ)
```

```python
# 5. Check if a number is an Armstrong number.

n = int(input("Enter the number"))
arm = 0
temp = n
while temp>0:
    digit = temp%10
    arm += (digit*digit*digit)
    temp //= 10
if arm == n:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# 6. Check if a number is a perfect number.

n = int(input("Enter the number"))
per = 1
for i in range(1,n+1):
    if i*i==n:
        print(i,"is the perfect square of",n)
        break
```

```python
# 7. Print all prime numbers between 1 and 100.

for num in range(2,101):
    prime = True
    for i in range(2,num):
        if num%i==0:
            prime = False
            break
    if prime:
        print(num,end=" ")
```

```python
# 8. Check if a number is prime or not.

n = int(input("Enter the number"))
prime = True
for i in range(2, int(n**0.5) +1):
    if n%i==0:
        prime = False
        break
if prime and n>1:
    print(n,"is a Prime Number")
else:
    print(n,"is not a Prime Number")
```

```python
# 9. Print Fibonacci series up to n terms.
# F(n) = F(n-1) + F(n-2)

n = int(input("Enter the number"))
a = 0
b = 1
for i in range(n):
    print(a,end=" ")
    a, b = b, a+b
```

```python
# 10. Print sum of first n terms of Fibonacci series.

n = int(input("Enter the number"))
a = 0
b = 1
summ = 0
for i in range(n):
    summ += a
    a, b = b, a+b
print(summ)
```

```python
# 5. Find LCM of two numbers using loops.

a = int(input("Enter the first number"))
b = int(input("Enter the second number"))
lcm = max(a,b)
while True:
    if lcm%a==0 and lcm%b==0:
        print(lcm)
        break
    lcm+=1
```

```python
# 8. Check if a number is a strong number (sum of factorials of digits = number).

a = int(input("Enter the number"))
summ = 0
temp = a
while temp>0:
    digit = temp % 10
    fact = 1
    for i in range(1,digit+1):       
        fact *= i
    summ += fact
    temp //= 10
if summ == a:
    print(f"Number {a} is a Strong Number")
else:
    print(f"Number {a} is not a Strong Number")
```

```python
# 9. Print first n terms of an arithmetic progression (a, d).

a = int(input("Enter the first term (a): "))
d = int(input("Enter the common difference (d): "))
n = int(input("Enter number of terms (n): "))

print("Arithmetic Progression:")

for i in range(n):
    term = a + i * d
    print(term)
```

```python
# 10. Print first n terms of a geometric progression (a, r).

a = int(input("Enter the first term (a): "))
r = int(input("Enter the common ratio (r): "))
n = int(input("Enter number of terms (n): "))

print("Geometric Progression:")

for i in range(n):
    term = a * (r ** i)
    print(term)
```

### Level 5: Logical Loop Combinations

```python
# 1. Print all numbers whose sum of digits is even (1–100).

for i in range(1,100):
    if (i%10 + i//10) %2 == 0:
        print(i)
```

```python
# 2. Count how many numbers between 1–500 are divisible by 7 but not by 5.

for i in range(1,500):
    if i%7==0 and i%5!=0:
        print(i)
```

```python
# 3. Print all numbers that are palindromes between 1–500.

for i in range(1,500):
    temp = i
    pal = 0
    while temp>0:
        digit = temp%10
        pal = pal*10 + digit
        temp //= 10
    if i == pal:
        print(i)
```

```python
# 4. Print numbers between 1–100 whose digits add up to a multiple of 3.

for i in range(1,100):
    temp = i
    add = 0
    while temp>0:
        digit = temp%10
        add = add + digit
        temp //= 10
    if add%3 == 0:
        print(i)
```

```python
# 5. Find the smallest and largest digit in a given number.

num = int(input("Enter a number :"))

small = 9
large = 0
temp = num
while temp > 0:
    digit = temp % 10 
    if digit < small:
        small = digit
    if digit > large:
        large = digit  
    temp //= 10

print("Smallest digit :", small)
print("Largest digit :", large)
    
```

```python
# 6. Print all numbers from 1–n whose binary representation has an even number of 1s.

n = int(input("Enter a number: "))
for i in range(1, n + 1):
    binary = bin(i)[2:]       
    ones = binary.count("1")  
    if ones % 2 == 0:         
        print(i, "->", binary)
```

```python
# 8. Print factorial of each number from 1 to n.

num = int(input("Enter a number :"))
fact = 1
for i in range(1,num+1):
    fact = i * fact
    print(f"Factorial of {i} is {fact}")
```

```python
# 9. Print the sum of all odd digits and even digits separately in a number.

num = int(input("Enter a number: "))
odd_sum = 0
even_sum = 0
temp = num
while temp > 0:
    digit = temp % 10   
    if digit % 2 == 0:
        even_sum += digit
    else:               
        odd_sum += digit
    temp //= 10         

print("Sum of even digits :", even_sum)
print("Sum of odd digits  :", odd_sum)
```

```python
# 10. Take 5 numbers as input. If the user enters 0, skip it using continue. 
# At the end, print the sum of all non-zero numbers entered.

total = 0
for i in range(5):
    num = int(input("Enter a number: "))
    if num == 0:    
        continue    
    total += num     

print("Sum of non-zero numbers:", total)
```
