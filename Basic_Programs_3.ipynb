## PHASE 3 — RECURSION (THINKING IN SELF-REFERENCE)

### Level 1: Foundation of Recursion (Base + Recursive Case)

```python
# 1. Print numbers from 1 to n using recursion.

def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

print_numbers(5)
```

```python
# 2. Print numbers from n down to 1 using recursion.

def print_numbers(n):
    if n == 0:
        return
    print(n)
    print_numbers(n - 1)
    
print_numbers(5)
```

```python
# 3. Print only even numbers from 1 to n recursively.

def even(n):
    if n == 0:
        return
    
    if n % 2 == 0:
        print(n)   
    even(n - 1)

even(10)
```

```python
# 4. Print only odd numbers from 1 to n recursively.

def odd(n):
    if n == 0:
        return
    
    if n % 2 == 1:
        print(n)   
    odd(n - 1)

odd(10)
```

```python
# 5. Print sum of first n natural numbers recursively.

def sum_n(n):
    if n == 0:       
        return 0
    return n + sum_n(n - 1)


sum_n(5)
```

```python
# 6. Print factorial of a number recursively.

def fact(n):
    if n==0:
        return True
    return n*fact(n-1)

fact(5)
```

```python
# 7. Calculate power of a number (xⁿ) using recursion.

def power(x, n):
    if n == 0:     
        return 1
    return x * power(x, n - 1)


power(2,5)
```

```python
# 8. Find nth Fibonacci number recursively.

def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

fib(6)
```

```python
# 9. Print Fibonacci series up to n terms recursively.

def fib(n, a=0, b=1):
    if n == 0:
        return
    print(a)
    fib(n - 1, b, a + b)


fib(6)
```

```python
# 10. Find sum of digits of a number recursively.

def digit(n):
    if n==0:
        return 0
    return (n % 10) + digit(n // 10)

digit(52)
```

### Level 2: Number-based Recursive Thinking

```python
# 1. Count the number of digits in a number recursively.

def num(n):
    if n < 10:
        return 1
    return 1 + num(n//10)
    
num(12345)
```

```python
# 2. Reverse a number recursively.

def rev(n,r=0):
    if n == 0:
        return r
    return rev(n//10, r*10 + n%10)
    
rev(123)
```

```python
# 3. Check if a number is a palindrome using recursion.

def reverse(n,rev=0):
    if n == 0:
        return rev
    return reverse(n//10, rev*10 + n%10)
def pal(n):
    return n==reverse(n)
        
n = int(input("Enter the number :"))
print("Palindrome" if pal(n) else "Not Palindrome")
```

```python
# 4. Find product of digits of a number recursively.

def prod(n):
    if n<10:
        return n
    return n%10 * prod(n//10)

prod(123)
```

```python
# 5. Find GCD (HCF) of two numbers using Euclid’s algorithm recursively.

def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

x = 48
y = 18

print(gcd(x, y))
```

```python
# 6. Convert a number to binary recursively.

def to_binary(n):
    if n == 0:
        return
    to_binary(n // 2)
    print(n % 2, end="")

num = 13

if num == 0:
    print(0)
else:
    to_binary(num)
```

```python
# 7. Print digits of a number in words recursively (e.g., 123 → “one two three”).

def words(n):
    word = ["zero", "one", "two", "three", "four",
             "five", "six", "seven", "eight", "nine"]

    if n == 0:
        return

    words(n // 10)
    print(word[n % 10], end=" ")


num = int(input("Enter the number"))
if num==0:
    print('Zero')
else:    
    words(num)
```

```python
# 8. Calculate the sum of first n even numbers recursively.

def sum_even(n):
    if n == 0:
        return 0
    return 2 * n + sum_even(n - 1)


n = 5   
print(sum_even(n))
```

```python
# 9. Calculate the sum of first n odd numbers recursively.

def sum_odd(n):
    if n == 0:
        return 0
    return (2 * n-1) + sum_odd(n - 1)


n = 5   
print(sum_odd(n))
```

```python
# 10. Find nCr (Combination formula) recursively using Pascal’s relation.

def nCr(n, r):
    if r == 0 or r == n:
        return 1
    return nCr(n - 1, r - 1) + nCr(n - 1, r)

n = 6
r = 3
print(nCr(n, r))
```

### Level 3: Pattern & Printing Problems

```python
# 1. Print a line of n stars recursively.

def star(n):
    if n==0:
        return
    star(n-1)
    print("*",end=" ")
star(5)
```

```python
# 2. Print a square of stars recursively (n×n).

def star(n,width=None):
    if width is None:
        width = n
    if n == 0:
        return
    print('*' * width)
    star(n - 1, width)
    
star(3)
```

```python
# 3. Print a triangle of stars recursively (top-down).

def triangle(n):
    if n==0:
        return
    triangle(n-1)
    print('*' * n)

triangle(5)    

print()

def triangle(n, width=None):
    if width is None:
        width = n
    if n == 0:
        return
    triangle(n - 1, width)
    print(' ' * (width - n) + '*' * n)

triangle(5)

print()

def pyramid(n, width=None):
    if width is None:
        width = n
    if n == 0:
        return
    pyramid(n - 1, width)
    print(' ' * (width - n) + '*' * (2 * n - 1))

pyramid(5)
```

```python
# 4. Print a triangle of stars recursively (bottom-up).

def triangle(n):
    if n==0:
        return
    print('*' * n)
    triangle(n-1)
    
triangle(5)
```

```python
# 5. Print pattern of numbers recursively (1 to n each row).

def print_numbers(n, i=1):
    if i > n:
        return
    print(i, end=' ')
    print_numbers(n, i + 1)

def pattern(rows, n):
    if rows == 0:
        return
    print_numbers(n)
    print()
    pattern(rows - 1, n)

pattern(5, 3)

print()

def row_numbers(i):
    if i == 0:
        return
    row_numbers(i - 1)
    print(i, end=' ')

def pattern(n):
    if n == 0:
        return
    pattern(n - 1)
    row_numbers(n)
    print()

pattern(5)

print()

def pattern(n, curr=1):
    if n == 0:
        return

    if curr > n:
        print()
        pattern(n - 1, 1)
    else:
        print(curr, end=' ')
        pattern(n, curr + 1)

pattern(5)
```

```python
# 6. Print reverse triangle pattern recursively.

def reverse_triangle(n):
    if n == 0:
        return
    print('*' * n)
    reverse_triangle(n - 1)

reverse_triangle(5)
```

```python
# 7. Print multiplication table of n recursively.

def table(n, i=1):
    if i > 10:
        return
    print(f"{n} x {i} = {n * i}")
    table(n, i + 1)

table(5)
```

```python
# 8. Print numbers in increasing and decreasing order in same function.

def inc_dec(n, curr=1):
    if curr > n:
        return
    print(curr, end=' ')
    inc_dec(n, curr + 1)
    if curr != n:
        print(curr, end=' ')

inc_dec(7)
```

```python
# 9. Print sum of series 1 + 2 + 3 + ... + n recursively and display each step.

def sum_series(n):
    if n == 0:
        return 0
    return n + sum_series(n - 1)

print(sum_series(5))

print()

def sum_series(n):
    if n == 0:
        print("Reached base case: sum = 0")
        return 0

    print(f"Calling sum_series({n-1}) to add {n}")
    result = n + sum_series(n - 1)
    print(f"Returning: {n} + sum_series({n-1}) = {result}")
    return result


n = 5
total = sum_series(n)
print("\nFinal Sum =", total)
```

```python
# 10. Print pattern of characters (A, AB, ABC, ...) recursively.

def char_pattern(n):
    if n == 0:
        return
    char_pattern(n - 1)
    print(''.join(chr(ord('A') + i) for i in range(n)))

char_pattern(3)

print()

def print_chars(n, i=0):
    if i == n:
        return
    print(chr(ord('A') + i), end='')
    print_chars(n, i + 1)

def char_pattern(n):
    if n == 0:
        return
    char_pattern(n - 1)
    print_chars(n)
    print()

char_pattern(5)
```

### Level 4: String-based Recursion

```python
# 1. Reverse a string using recursion.

def reverse(s):
    if len(s)<=1:
        return s
    return reverse(s[1:])+s[0]
    
print(reverse("hello"))
```

```python
# 2. Check if a string is palindrome using recursion.

def rev(s):
    if len(s)<=1:
        return s
    return rev(s[1:])+s[0]

def pal(s):
    if s==rev(s):
        print(f'"{s}" is a Palindrome')
    else:
        print(f'"{s}" is Not a Palindrome')

pal('madam')
```

```python
# 3. Count vowels in a string recursively.

def vowel(s):
    if s == '':
        return 0
    if s[0] in 'aeiouAEIOU':
        return 1 + vowel(s[1:])
    else:
        return vowel(s[1:])
        
print(vowel('hello how are you'))
```

```python
# 4. Remove all spaces from a string recursively.

def remove_spaces(s):
    if s == "":
        return ""
    if s[0] == " ":
        return remove_spaces(s[1:])
    else:
        return s[0] + remove_spaces(s[1:])

print(remove_spaces('hello sir how are you'))
```

```python
# 5. Replace all occurrences of a character (say ‘a’ → ‘x’) recursively.

def replace(s):
    if s == "":
        return ""
    if s[0] == 'a':
        return 'x' + replace(s[1:])
    else:
        return s[0] + replace(s[1:])

replace('maaadaaaam')
```

```python
# 6. Remove all occurrences of a character from a string recursively.

def remove_char(s, ch):
    if s == "":
        return ""
    if s[0] == ch:
        return remove_char(s[1:], ch)
    else:
        return s[0] + remove_char(s[1:], ch)

remove_char("maam",'a')
```

```python
# 7. Print all characters of a string one by one recursively.

def print_chars(s):
    if s == "":
        return
    print(s[0])
    print_chars(s[1:])

print_chars('madam')
```

```python
# 8. Print the string in reverse order recursively (without using loops).

def print_reverse(s):
    if s == "":
        return
    print_reverse(s[1:])
    print(s[0], end="")

print_reverse('hello')
```

```python
# 9. Convert a string to uppercase recursively.

def to_uppercase(s):
    if s == "":
        return ""
    return s[0].upper() + to_uppercase(s[1:])

to_uppercase('hello')
```

```python
# 10. Count consonants and vowels separately using recursion.

def count_vowels_consonants(s):
    if s == "":
        return 0, 0 
    vowel = "aeiouAEIOU"
    v, c = count_vowels_consonants(s[1:])
    if s[0].isalpha():
        if s[0] in vowel:
            return v + 1, c
        else:
            return v, c + 1
    return v,c
    
vowels, consonants = count_vowels_consonants("hello how are you bro")
print(f"{vowels}_Vowels, {consonants}_Consonants")
```
