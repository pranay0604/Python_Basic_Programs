### Category 2 : Counting & Character Analysis

```python
# 1. Count how many vowels and consonants are in a string.

s = "Hello World"
conso = 0
vowel = 0
for i in s:
    if i in "aeiouAEIOU":
        vowel +=1
    else:
        conso +=1
print(f"Vowels : {vowel}\nConsonants : {conso}")
```

```python
# 2. Count the number of digits, letters, and special characters in a string.

s = "Hello123@#World!"
letters = 0
digits = 0
special = 0
for ch in s:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1
    else:
        special += 1   
print("Letters:", letters)
print("Digits:", digits)
print("Special characters:", special)
```

```python
# 3. Count how many uppercase and lowercase letters a string has.

s = "Hello World"
upper = 0
lower = 0
for ch in s:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
print("Uppercase:", upper)
print("Lowercase:", lower)
```

```python
# 4. Find the frequency of each character in a string (without using a map).

s = "Hello World"
ans = ""

for ch in s:
    if ch not in ans:
        print(ch, ":", s.count(ch))
        ans += ch
```

```python
# 5. Count how many spaces are there in a sentence.

s = "Hello World Welcome to Python"
count = 0
for i in s:
    if i == " ":
        count+=1
print(count)
```

```python
# 6. Count how many times a given character appears in a string.

s = "Hello World"
ch = "l"
count = 0
for c in s:
    if c == ch:
        count += 1
print(f"'{ch}' appears {count} times")
```

```python
# 7. Count how many alphabets are before ‘m’ and after ‘m’ in a given string.

s = "Hello World"
before_m = 0
after_m = 0
for ch in s:
    if ch.isalpha():             
        c = ch.lower()           
        if c < 'm':
            before_m += 1
        elif c > 'm':
            after_m += 1
print("Before m:", before_m)
print("After m:", after_m)
```

```python
# 8. Count how many substrings start and end with the same character (simple logic).

s = "abcab"
total = 0
for ch in set(s):          
    f = s.count(ch)        
    total += f * (f + 1) // 2
print(total)
```

```python
# 9. Print how many words start with a vowel in a sentence.

s = "Hello everyone welcome to India and Python Course"
vowels = "aeiouAEIOU"
count = 0
words = s.split()
for i in words:
    if i[0] in vowels:
        count += 1
print("Words starting with a vowel:", count)
```

```python
# 10. Count how many words end with ‘s’.

s = "Hello everyone welcome to Full Stacks Data Science and Python Courses"
count = 0
words = s.split()
for i in words:
    if i[-1] == 's':
        count += 1
print("Words ending with s letter :", count)
```

### Category 3 : Reversing & Palindromic Thinking

```python
# 1. Reverse a string without using built-in reverse.

s = "Hello World"
rev = ''
for i in s:
    rev = i + rev
print(rev)

print()
rev1 = ''
for j in range(len(s) - 1, -1, -1):
    rev1 += s[j]
print(rev)

print()
print("".join(list(reversed(s))))
```

```python
# 2. Reverse each word in a sentence.

s = "Hello World from Python"
words = s.split()
result = ""
for w in words:
    rev = ""
    for ch in w:
        rev = ch + rev
    result += rev + " "
print(result.strip())
```

```python
# 3. Reverse the order of words in a sentence.

s = "Hello World from Python"
words = s.split()
result = ""
for i in range(len(words) - 1, -1, -1):
    result += words[i] + " "
print(result.strip())

print()
print(" ".join(words[::-1]))
```

```python
# 4. Check whether a string is a palindrome.

s = "madam"
s = s.lower()
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

print()
s1 = "Hello"
rev = ""
for ch in s1:
    rev = ch + rev
if s1 == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# 5. Check if two strings are the reverse of each other.

s1 = "abcd"
s2 = "dcba"
if s1 == s2[::-1]:
    print("Reverse of each other")
else:
    print("Not reverse")
```

```python
# 6. Print the middle character(s) of a string.

s = "Hello World WeLcome to Python"
n = len(s)
if n % 2 == 1:
    print("Middle character :", s[n // 2])
else:
    print("Middle character :", s[n // 2 - 1] + s[n // 2])
```

```python
# 7. Print the second half of the string in reverse.

s = "Hello World"
n = len(s)
second_half = s[n//2:]    
rev = ""
for ch in second_half:
    rev = ch + rev          
print(rev.strip())
```

```python
# 8. Remove the first and last character and print the remaining string.

s = "Hello World"
print(s[1:-1])

print()
result = ""
for i in range(1, len(s) - 1):
    result += s[i]
print(result)
```

```python
# 9. Reverse only characters, keeping digits in place.

s = "a1b2c3"
letters = []
for ch in s:
    if ch.isalpha():
        letters.append(ch)
result = ""
for ch in s:
    if ch.isalpha():
        result += letters.pop()  
    else:
        result += ch
print(result)
```

```python
# 10. Reverse string but skip spaces.

s = "Hello World"
chars = []
for ch in s:
    if ch != " ":
        chars.append(ch)
result = ""
k = len(chars) - 1
for ch in s:
    if ch == " ":
        result += " "
    else:
        result += chars[k]
        k -= 1
print(result)
```

### Category 4 : Character & Word Manipulation

```python
# 1. Remove all vowels from a string.

s = "Hello World"
vowel = "aeiouAEIOU"
ans = ''
for i in s:
    if i in vowel:
        pass
    else:
        ans += i
print(ans)
```

```python
# 2. Remove all spaces from a string.

s = "Hello World"
space = " "
ans = ''
for i in s:
    if i in space:
        pass
    else:
        ans += i
print(ans)

print()
print(s.replace(" ", ""))
```

```python
# 3. Replace all vowels with ‘*’.

s = "Hello World"
vowel = "aeiouAEIOU"
ans = ''
for i in s:
    if i in vowel:
        ans += '*'
    else:
        ans += i
print(ans)
```

```python
# 4. Replace all spaces with ‘_'.

s = "Hello World"
space = " "
ans = ''
for i in s:
    if i in space:
        ans += '_'
    else:
        ans += i
print(ans)

print()
print(s.replace(" ", "_"))
```

```python
# 5. Print the string after removing all digits.

s = "He110 W0r1d 2026"
ans = ""
for ch in s:
    if not ch.isdigit():
        ans += ch
print(ans)

print()
s1 = "He110 W0r1d 2026"
ans1 = ''
for ch in s1:
    if ch.isdigit():
        continue
    ans1 += ch
print(ans1)
```

```python
# 6. Remove duplicate characters from a string.

s = "Hello World"
ans = ""
for ch in s:
    if ch not in ans:
        ans += ch
print(ans)

print()
print("".join(dict.fromkeys(s)))
```

```python
# 7. Keep only the first occurrence of each character.

s = "Hello World"
ans = ""
for ch in s:
    if ch != " " and ch not in ans:
        ans += ch
print(ans)
```

```python
# 8. Remove consecutive duplicate characters (e.g.“aaabb” → “ab”).

s = "aaabbccc"
ans = ""
for ch in s:
    if ans == "" or ch != ans[-1]:
        ans += ch
print(ans)
```

```python
# 9. Swap case: uppercase → lowercase and lowercase → uppercase.

s = "Hello World"
ans = ""
for ch in s:
    if ch.isupper():
        ans += ch.lower()
    elif ch.islower():
        ans += ch.upper()
    else:
        ans += ch
print(ans)
```

```python
# 10. Shift each character by 1 (e.g., “abc" -> "bcd").

s = "abc"
ans = ""
for ch in s:
    ans += chr(ord(ch) + 1)
print(ans)

print()
s = "xyz"
ans = ""
for ch in s:
    if 'a' <= ch <= 'z':
        ans += chr((ord(ch) - ord('a') + 1) % 26 + ord('a'))
    elif 'A' <= ch <= 'Z':
        ans += chr((ord(ch) - ord('A') + 1) % 26 + ord('A'))
    else:
        ans += ch
print(ans)
```

### Category 5 : Word-level Thinking

```python
# 1. Print each word of a sentence on a new line.

s = "Hello World"
s = s.split()
for i in s:
    print(i)
```

```python
# 2. Count how many words have even length.

s = "Hello World Welcome to Python Course"
s = s.split()
count = 0
for i in s:
    if len(i)%2==0:
        count+=1
print(count)
```

```python
# 3. Find the longest word in a sentence.

s = "Hello World Welcome to Python Course"
words = s.split()
longest = words[0]
for w in words:
    if len(w) > len(longest):
        longest = w
print("Longest word:", longest)
print("Length:", len(longest))
```

```python
# 4. Find the shortest word in a sentence.

s = "Hello World Welcome to Python Course"
words = s.split()
smallest = words[0]
for w in words:
    if len(w) < len(smallest):
        smallest = w
print("Smallest word:", smallest)
print("Length:", len(smallest))
```

```python
# 5. Swap first and last words in a sentence.

s = "Hello World Welcome to Python Course"
words = s.split()
words[0], words[-1] = words[-1], words[0]
print(" ".join(words))
```

```python
# 6. Print all words that start and end with the same letter.

s = "Hello World Welcome to Python Course begineer level at noon"
words = s.split()
for w in words:
    if len(w) > 0 and w[0].lower() == w[-1].lower():
        print(w)
```

```python
# 7. Count how many words contain the letter ‘a’.

s = "I am learning Java and Python at home"
words = s.split()
count = 0
for i in words:
    if 'a' in i.lower():
        count+=1
        print(i)
print(count)
```

```python
# 8. Capitalize the first letter of each word.

s = "I am learning java and python at home"
ans = s.title()
print(ans)

print()
words = s.split()
ans1 = ""
for w in words:
    ans1 += w[0].upper() + w[1:] + " "

print(ans1.strip())
```

```python
# 9. Print the sentence in title case (first letter capital, rest lowercase).

s = 'hello world'
ans = s.title()
print(ans)

print()
ans1 = s.capitalize()
print(ans1)
```
