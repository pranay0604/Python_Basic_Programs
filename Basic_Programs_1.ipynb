```python
# 1. Take a 3-digit number and check if all digits are distinct.

num = int(input("Enter a 3-digit number: "))

a = num // 100       
b = (num // 10) % 10 
c = num % 10        

digits = [a, b, c]
duplicate_count = 0


for i in range(3):
    for j in range(i + 1, 3):
        if digits[i] == digits[j]:
            duplicate_count += 1


if duplicate_count == 0:
    print("All digits are distinct.")
else:
    print("Digits are NOT distinct.")
```

```python
num = int(input("Enter a number: "))

low = 1
high = num
is_square = False

while low <= high:
    mid = (low + high) // 2
    square = mid * mid

    if square == num:
        is_square = True
        break
    elif square < num:
        low = mid + 1
    else:
        high = mid - 1

if is_square:
    print("Perfect square")
else:
    print("Not a perfect square")
```

```python
# 8. Take an integer (1–9999) and check if the sum of its digits is greater than the product of its digits.

num = int(input("Enter the number between (1-9999)"))
temp = num
summ = 0
prod = 1
while temp > 0:
    digit = temp % 10
    summ += digit
    prod *= digit
    temp //= 10
print("Summation is ",summ)
print("Product is ",prod)
if summ > prod:
    print("Sum of Digits is Greater than the Product of its Digits")
else:
    print("Sum of Digits is not Greater than the Product of its Digits")
```

```python
# 9. Take two dates (day and month) and determine which one comes first in the calendar.

d1 = int(input("Enter day of first date: "))
m1 = int(input("Enter month of first date: "))
d2 = int(input("Enter day of second date: "))
m2 = int(input("Enter month of second date: "))
if m1 < m2:                                          
    print("First date comes first:", d1, "/", m1)
elif m1 > m2:
    print("Second date comes first:", d2, "/", m2)
else:                                                 
    if d1 < d2:
        print("First date comes first:", d1, "/", m1)
    elif d1 > d2:
        print("Second date comes first:", d2, "/", m2)
    else:
        print("Both dates are the same")
```

```python
# 10. Take a year and print the corresponding century (e.g.,“19th century”,“20th century”)

year = int(input("Enter a year: "))
century = (year - 1) // 100 + 1 
if century % 10 == 1 and century % 100 != 11:
    suffix = "st"
elif century % 10 == 2 and century % 100 != 12:
    suffix = "nd"
elif century % 10 == 3 and century % 100 != 13:
    suffix = "rd"
else:
    suffix = "th"
print(f"{century}{suffix} century")
```
