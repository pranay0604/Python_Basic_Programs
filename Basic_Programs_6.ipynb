### Category 1 : Number-Based Logical Combinations

```python
# 1. Print all numbers between 1 and N that are divisible by both 3 and 5.

n = int(input("Enter the Nth term :"))
for i in range(1,n+1):
    if i%3==0 and i%5==0:
        print(i)
```

```python
# 2. Find the sum of digits of a number (use loop).

n = int(input("Enter the number :"))
summ = 0
while n > 0:
    digit = n % 10
    summ += digit
    n //= 10
print(summ)
```

```python
# 3. Check if a number is an Armstrong number.

n = int(input("Enter the number :"))
summ = 0
digits = 0
temp = n
while temp>0:
    digits+=1
    temp//=10
temp = n
while temp > 0:
    digit = temp%10
    summ += digit ** digits
    temp//=10
if summ == n:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# 4. Print all Armstrong numbers between 1 and 1000.

for n in range(1, 1001):
    temp = n
    digits = 0
    summ = 0
    while temp > 0:
        digits += 1
        temp //= 10
    temp = n
    while temp > 0:
        digit = temp % 10
        summ += digit ** digits
        temp //= 10
    if summ == n:
        print(n)
```

```python
# 5. Find the factorial of a number using recursion.

def fact(n):
    if n==0 or n==1:
        return 1
    return n * fact(n-1)

fact(0)
```

```python
# 6. Count how many even digits a number contains.

n = int(input("Enter the number :"))
count = 0
while n > 0:
    digit = n % 10
    if digit%2==0:
        count += 1
    n //= 10  
print(count)
```

```python
# 7. Print all prime numbers between 1 and N.

n = int(input("Enter the value of N: "))
for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)

print()

n = int(input("Enter N term : "))
for num in range(2, n + 1):
    count = 0
    for i in range(1, num + 1):
        if num % i == 0:
            count += 1
    if count == 2:
        print(num)
```

```python
# 8. Print the reverse of a number (123 → 321).

n = int(input("Enter the number: "))
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10
print(rev)
```

```python
# 9. Check if a number is palindrome (121 → true).

n = int(input("Enter the number: "))
rev=0
temp = n
while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10
if n == rev:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
# 10. Check if a number is perfect (sum of factors equals number).

n = int(input("Enter the number: "))
summ = 0
for i in range(1, n):
    if n % i == 0:
        summ += i
if summ == n:
    print("Perfect number")
else:
    print("Not a Perfect number")
```

### Category 2 : String + Logic Mix

```python
# 1. Check if two strings are anagrams (without using collections).

s1 = input("Enter first string: ").replace(" ", "").lower()
s2 = input("Enter second string: ").replace(" ", "").lower()
if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not anagrams")

print()

s1 = input("Enter first string: ").replace(" ", "").lower()
s2 = input("Enter second string: ").replace(" ", "").lower()
if len(s1) != len(s2):
    print("Not anagrams")
else:
    for ch in s1:
        if s1.count(ch) != s2.count(ch):
            print("Not anagrams")
            break
    else:
        print("Anagrams")
```

```python
# 2. Count vowels in each word of a sentence.

sentence = input("Enter a sentence: ").lower()
vowels = "aeiou"
words = sentence.split()
for word in words:
    count = 0
    for ch in word:
        if ch in vowels:
            count += 1
    print(word, ":", count)
```

```python
# 3. Reverse words in a string if their length is even.

sentence = input("Enter a sentence: ")
words = sentence.split()
result = []
for word in words:
    if len(word) % 2 == 0:
        result.append(word[::-1])
    else:
        result.append(word)
print(" ".join(result))
```

```python
# 4. Replace every vowel in a string with its position (a=1, e=2...).

s = input("Enter a string: ").lower()
vowels = "aeiou"
result = ""
for ch in s:
    if ch in vowels:
        result += str(vowels.index(ch) + 1)
    else:
        result += ch
print(result)
```

```python
# 5. Print characters that appear more than once (without map).

s = input("Enter a string: ")
ans = ""

for ch in s:
    if s.count(ch) > 1 and ch not in ans:
        print(ch)
        ans += ch
else:
    print("All characters are unique")
```

```python
# 6. Count words that start and end with the same letter.

sentence = input("Enter a sentence: ").lower()
words = sentence.split()
count = 0
for word in words:
    if len(word) > 0 and word[0] == word[-1]:
        count += 1
print("Count:", count)
```

```python
# 7. Toggle case for every alternate word in a sentence.

sentence = input("Enter a sentence: ")

words = sentence.split()
result = []
for i in range(len(words)):
    if i % 2 == 1: 
        result.append(words[i].swapcase())
    else:
        result.append(words[i])
print(" ".join(result))
```

```python
# 8. Check if two strings are rotations of each other.

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
if len(s1) == len(s2) and s2 in (s1 + s1):
    print("Strings are rotations of each other")
else:
    print("Strings are not rotations of each other")
```

```python
# 9. Find the word with maximum vowels in a sentence.

sentence = input("Enter a sentence: ").lower()
vowels = "aeiou"
words = sentence.split()
max_count = 0
max_word = ""
for word in words:
    count = 0
    for ch in word:
        if ch in vowels:
            count += 1
    if count > max_count:
        max_count = count
        max_word = word

print("Word with maximum vowels:", max_word)
print("Number of vowels:", max_count)
```

```python
# 10. Remove duplicate words from a sentence.

sentence = input("Enter a sentence: ")
words = sentence.split()
result = []
for word in words:
    if word not in result:
        result.append(word)
print(" ".join(result))
```

### Category 3 : Array + Looping Logic

```python
# 1. Find the maximum and minimum element in an array.

arr = list(map(int, input("Enter array elements: ").split()))
maximum = arr[0]
minimum = arr[0]
for i in arr:
    if i > maximum:
        maximum = i
    if i < minimum:
        minimum = i
print("Maximum element:", maximum)
print("Minimum element:", minimum)
```

```python
# 2. Count how many positive, negative, and zero elements are in an array.

arr = list(map(int, input("Enter array elements: ").split()))
pos = 0
neg = 0
zero = 0
for num in arr:
    if num > 0:
        pos += 1
    elif num < 0:
        neg += 1
    else:
        zero += 1
print("Positive numbers:", pos)
print("Negative numbers:", neg)
print("Zero's:", zero)
```

```python
# 3. Print all unique elements from an array.

arr = list(map(int, input("Enter array elements: ").split()))
for i in range(len(arr)):
    count = 0
    for j in range(len(arr)):
        if arr[i] == arr[j]:
            count += 1
    if count == 1:
        print(arr[i], end=" ")
```

```python
# 4. Reverse an array in-place.

arr = list(map(int, input("Enter array elements: ").split()))
start = 0
end = len(arr) - 1
while start < end:
    arr[start], arr[end] = arr[end], arr[start]
    start += 1
    end -= 1
print("Reversed array:", arr)
```

```python
# 5. Shift all zeros to the end of the array.

arr = list(map(int, input("Enter array elements: ").split()))
pos = 0 
for i in range(len(arr)):
    if arr[i] != 0:
        arr[pos], arr[i] = arr[i], arr[pos]
        pos += 1
print("After shifting zeros:", arr)
```

```python
# 6. Count how many elements are even at an even index.

arr = list(map(int, input("Enter array elements: ").split()))
count = 0
for i in range(len(arr)):
    if i % 2 == 0 and arr[i] % 2 == 0:
        count += 1
print("Count:", count)
```

```python
# 7. Merge two arrays into one.

arr1 = list(map(int, input("Enter first array: ").split()))
arr2 = list(map(int, input("Enter second array: ").split()))
ans = []
for x in arr1:
    ans.append(x)
for x in arr2:
    ans.append(x)
print("Merged array:", ans)
```

```python
# 8. Find the second largest element in an array.

arr = list(map(int, input("Enter array elements: ").split()))
largest = second = float('-inf')
for num in arr:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num
if second == float('-inf'):
    print("No second largest element")
else:
    print("Second largest element:", second)
```

```python
# 9. Rotate an array by one position to the right.

arr = list(map(int, input("Enter array elements: ").split()))
last = arr[-1]
for i in range(len(arr) - 1, 0, -1):
    arr[i] = arr[i - 1]
arr[0] = last
print("Rotated array:", arr)
```

```python
# 10. Find the sum of all elements at odd indices.

arr = list(map(int, input("Enter array elements: ").split()))
summ = 0
for i in range(len(arr)):
    if i % 2 == 1:
        summ += arr[i]
print("Sum of elements at odd indices:", summ)
```

### Category 4 : Nested Logic & Pattern Flow

```python
# 1. Print a multiplication table in a formatted grid (10x10).

n = int(input("Enter N: "))
for i in range(1, n + 1):
    for j in range(1, n + 1):
        print(f"{i*j:5}", end=" ")
    print()
```

```python
# 2. Print all pairs in an array whose sum equals a given number.

arr = list(map(int, input("Enter array elements: ").split()))
target = int(input("Enter target sum: "))
n = len(arr)
for i in range(n):
    for j in range(i + 1, n):
        if arr[i] + arr[j] == target:
            print(arr[i], arr[j])
```

```python
# 3. Print all subarrays of a given array.

arr = list(map(int, input("Enter array elements: ").split()))
n = len(arr)
for i in range(n):
    for j in range(i, n):
        for k in range(i, j + 1):
            print(arr[k], end=" ")
        print()
```

```python
# 4. Check if an array is sorted (ascending or descending).

arr = list(map(int, input("Enter array elements: ").split()))
asc = True
desc = True
for i in range(len(arr) - 1):
    if arr[i] > arr[i + 1]:
        asc = False
    if arr[i] < arr[i + 1]:
        desc = False
if asc:
    print("Array is sorted in ascending order")
elif desc:
    print("Array is sorted in descending order")
else:
    print("Array is not sorted")
```

```python
# 5. Count how many times a number appears consecutively in an array.

arr = list(map(int, input("Enter array elements: ").split()))
x = int(input("Enter the number: "))
count = 0
max_count = 0
for num in arr:
    if num == x:
        count += 1
        if count > max_count:
            max_count = count
    else:
        count = 0
print("Maximum consecutive count:", max_count)
```

```python
# 6. Find all pairs of characters in a string that are the same (nested loop).

s = input("Enter a string: ")
n = len(s)
for i in range(n):
    for j in range(i + 1, n):
        if s[i] == s[j]:
            print(f"({s[i]}, {s[j]}) at indices ({i}, {j})")
```

```python
# 7. Print pattern of increasing characters (A, AB, ABC...).

n = int(input("Enter number of rows: "))
for i in range(1, n + 1):
    ch = 'A'
    for j in range(i):
        print(chr(ord(ch) + j), end="")
    print()
```

```python
# 8. Print Pascal’s triangle up to N rows.

n = int(input("Enter number of rows: "))
for i in range(n):
    num = 1
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```

```python
# 9. Generate Fibonacci series up to N using recursion.

def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

n = int(input("Enter number of terms: "))
for i in range(n):
    print(fib(i), end=" ")
```

```python
# 10. Print numbers in a spiral-like pattern (conceptual dry run).

n = int(input("Enter N: "))
arr = [[0] * n for _ in range(n)]
top = 0
bottom = n - 1
left = 0
right = n - 1
num = 1
while top <= bottom and left <= right:
    for i in range(left, right + 1):
        arr[top][i] = num
        num += 1
    top += 1

    for i in range(top, bottom + 1):
        arr[i][right] = num
        num += 1
    right -= 1

    for i in range(right, left - 1, -1):
        arr[bottom][i] = num
        num += 1
    bottom -= 1

    for i in range(bottom, top - 1, -1):
        arr[i][left] = num
        num += 1
    left += 1
print()
for row in arr:
    for val in row:
        print(f"{val:3}", end=" ")
    print()
```

### Category 5 : Applied Reasoning & Real-Life Logic

```python
# 1. Given marks of students, find how many passed (>= 40).

marks = list(map(int, input("Enter marks: ").split()))
count = 0
for m in marks:
    if m >= 40:
        count += 1
print("Number of students passed:", count)
```

```python
# 2. Take age inputs and count how many are adults, minors, seniors.

ages = list(map(int, input("Enter ages: ").split()))
minors = 0
adults = 0
seniors = 0
for age in ages:
    if age < 18:
        minors += 1
    elif age < 60:
        adults += 1
    else:
        seniors += 1
print("Minors:", minors)
print("Adults:", adults)
print("Seniors:", seniors)
```

```python
# 3. Validate a password (at least one uppercase, lowercase, digit, special char).

password = input("Enter password: ")
upper = lower = digit = special = 0
for ch in password:
    if ch.isupper():
        upper = 1
    elif ch.islower():
        lower = 1
    elif ch.isdigit():
        digit = 1
    else:
        special = 1
if upper and lower and digit and special:
    print("Valid password")
else:
    print("Invalid password")
```

```python
# 4. Simulate a simple calculator using switch-case.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operation (+, -, *, /): ")
match op:
    case "+":
        print("Result:", a + b)
    case "-":
        print("Result:", a - b)
    case "*":
        print("Result:", a * b)
    case "/":
        if b != 0:
            print("Result:", a / b)
        else:
            print("Division by zero not allowed")
    case _:
        print("Invalid operation")
```

```python
# 5. Count how many times a coin lands on heads/tails (use random).

import random
n = int(input("Enter number of tosses: "))
heads = 0
tails = 0
for i in range(n):
    toss = random.choice(["H", "T"])
    if toss == "H":
        heads += 1
    else:
        tails += 1
print("Heads:", heads)
print("Tails:", tails)
```

```python
# 9. Count how many prime numbers are there in an array.

arr = list(map(int, input("Enter array elements: ").split()))
prime_count = 0
for num in arr:
    if num > 1:
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
            prime_count += 1
print("Number of prime numbers:", prime_count)
```
